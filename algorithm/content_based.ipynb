Weight Calculation

```python
results = []
for username, user_restaurant1 in user_rating_dict.items():
    # Extract the corresponding user restaurant dictionary
    user_restaurant_dict1 = user_restaurant_dict[username]

    # Calculate the weighted values
    for key, rating in user_restaurant1.items():
        tmp = restaurant_feature_dict[key]
        for item in tmp:
            weight = 0.2
            for lower, upper in [(1.0, 1.9), (2.0, 2.9), (3.0, 3.0), (3.1, 4.0), (4.1, 5.0)]:
                if lower <= rating <= upper:
                    user_restaurant_dict1[item] += weight
                weight += 0.2

    # Normalize the values
    new_user_restaurant_dict1 = {
        key: value / len(restaurant_feature_dict) for key, value in user_restaurant_dict1.items()
    }

    # Append the result to the list
    results.append({
        "user_id": username,
        "calculated_value": new_user_restaurant_dict1
    })

# Convert the results to a dataframe
user_profile = pd.DataFrame(results)

# Display the dataframe
user_profile
```

Pearson Correlation

```python
from scipy.stats import pearsonr

# Step 1: Filter restaurant profiles based on test data
result_list = []
for user in user_restaurant_df['user_id'].unique():
    # Get restaurants associated with the user
    user_restaurants = user_restaurant_df[user_restaurant_df['user_id'] == user]['restaurant_id']
    restaurant_features = restaurant_profile[restaurant_profile['restaurant_id'].isin(user_restaurants)]
    
    # Step 2: Compute Pearson_Correlatsion for the user
    user_profile = user_profile_cs[user_profile_cs['user_id'] == user].iloc[:, 1:].values.flatten()  # Drop 'username' and convert to array
    
    # Calculate Pearson_Correlation between user and each restaurant
    for index, restaurant_row in restaurant_features.iterrows():
        restaurant_vector = restaurant_row.iloc[1:].values.flatten()  # Drop 'name' column and convert to array
        restaurant_vector = restaurant_vector.astype(float)
        distance = pearsonr(user_profile, restaurant_vector)  # Calculate Pearson_Correlation
        result_list.append({
            'user_id': user,
            'restaurant_id': restaurant_row['restaurant_id'],
            'Pearson_Correlation': distance
        })

# Create a DataFrame from the results
pearson_correlation_df = pd.DataFrame(result_list)
pearson_correlation_df
```

```python
pearson_correlation_df['Pearson_Coeff'] = pearson_correlation_df['Pearson_Correlation'].apply(lambda x: x[0])
pearson_correlation_df
```

```python
user_pcc_sort = pearson_correlation_df.groupby("user_id", group_keys=False).apply(lambda x: x.sort_values("Pearson_Coeff", ascending=False))

# Reset index if needed
user_pcc_sort = user_pcc_sort.reset_index(drop=True)

user_pcc_sort
```
